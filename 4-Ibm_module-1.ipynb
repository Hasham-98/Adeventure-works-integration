```python
# perform some transformation
```

```python
# The function to convert from string to float value
def str_to_num(number):
    try:
        num = float(number)    # data casting
        return num
    except:
        try:
            if "$" in number:
                number = number.replace('$','')
                num = float(number)  # data casting
                return num
        except:
            print(number)
            return number
```

```python
data[['Freight']]
```

```python
data['Freight']   = data['Freight'].apply(str_to_num)
```

```python
data['UnitPrice'] = data['UnitPrice'].apply(str_to_num)
```

```python
data['LineTotal'] = data['LineTotal'].apply(str_to_num)
```

```python
data['TaxAmt']    = data['TaxAmt'].apply(str_to_num)
```

```python
data['TotalDue']  = data['TotalDue'].apply(str_to_num)
```

```python
data['TotalDue']
```

```python
# convert from $ to uero
data['TotalDue'] = data['TotalDue']*1.08
data['TotalDue']
```
